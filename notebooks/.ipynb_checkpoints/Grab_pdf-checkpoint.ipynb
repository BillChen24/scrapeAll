```python
def get_links_on_iframe(driver):
    # Wait until the sub page pops up
    WebDriverWait(driver, 60).until(EC.presence_of_element_located((By.ID, "jbox-iframe")))
    driver.switch_to.frame(driver.find_element_by_id("jbox-iframe"))
    
    
    folder_title =  driver.find_element_by_xpath("//div[@class='infolistbot']//h3").text
    parent_file_html = driver.find_elements_by_xpath("//div[@class='detail-addfile']//a")
    
    files = [i.get_attribute("href") for i in parent_file_html]
    pdf_name = [i.get_attribute("title") for i in parent_file_html]
    file_dict = dict(zip(pdf_name, files))

    driver.switch_to.default_content()
    close_button = driver.find_element_by_xpath("//a[@class='jbox-close']")
    close_button.click()
    
    output = {folder_title:file_dict}
    print(output)
    return output
    
```

```python
def get_links_on_page(driver):
    myElem = WebDriverWait(driver, 60).until(EC.presence_of_element_located((By.CLASS_NAME, 'newslist')))
    print("Page is ready!")

    parentElement = driver.find_element_by_class_name("newslist")
    elementList = parentElement.find_elements_by_tag_name("a")
    output = []
    for i in elementList:
        i.click()
        output.append(get_links_on_iframe(driver))
    return output
    
```

```python
def get_links_on_all_pages(driver):
    next_button = driver.find_element_by_xpath("//a[text()='下一页']")
    output = []
    while next_button.get_attribute("class") !='disabled': 
        output.extend(get_links_on_page(driver))
        next_button.click()
        next_button = driver.find_element_by_xpath("//a[text()='下一页']")
    output.extend(get_links_on_page(driver))
    return output
    


       

```

```python
all_links = get_links_on_all_pages(driver)
```

```python
with open('file_links_chongqing.json', 'w') as f:
    json.dump(all_links, f)
```

```python
import os
os.mkdir('./ChongQing')
```

```python
for i in all_links:
    for folder_name, file_dict in i.items():
        os.mkdir('./ChongQing/'+folder_name)
        for file_name, url in file_dict.items():
            print(file_name, url)
            cur_pdf = requests.get(url, verify = False)
            
            with open('./ChongQing/'+folder_name+'/'+file_name, 'wb') as f:
                f.write(cur_pdf.content)
```
